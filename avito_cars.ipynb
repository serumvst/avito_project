```python
page = 1

while True:
    
    url = ("https://www.avito.ru/moskva_i_mo/avtomobili?cd=1&p=" + str(page)+ "&q=bmw")
    r = requests.get(url)
    html = BS(r.content, "html.parser")
    pages = html.select(".items-items-kAJAg > .iva-item-root-_lk9K")
    
    if (len(pages)):
        for el in pages:
            # Получаем всю информацию об автомобиле
            info = el.find("div", {"class" : "iva-item-title-py3i_" }).find("a").get("title")

            # Выводим всю информацию об автомобиле
            print(f'{info} \n')

            # Разделяем строку по запятым и пробелам
            parts = info.split(", ")

            # Получаем название автомобиля
            car_name = parts[0]

            print("Название:", car_name)

            # Получаем год выпуска
            car_age = parts[1]

            print("Год выпуска:", car_age)

            # Получаем километраж 
            car_km = parts[2]

            print("Километраж", car_km)

            # Получаем стоимость
            car_price = parts[-1]


            price_start = car_price.find("цена") + 5
            price_end = car_price.find("руб")
            price = car_price[price_start:price_end].strip()

            print("Стоимость:", price)

            print(f'\n')
    else:
        break
```

## <u>Импортируем_в_csv-файл</u>

```python
import requests
from bs4 import BeautifulSoup as BS
import csv
import time
```

```python
# Задаем настройки прокси
proxies = {
    'http': 'http://10.10.1.10:3128',
    'https': 'https://10.10.1.10:1080',
}

with open('cars_info.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Название', 'Год выпуска', 'Километраж', 'Стоимость'])

    page = 1
    while True:
        try:
            url = f"https://www.avito.ru/moskva_i_mo/avtomobili?cd=1&p={page}&q=bmw"
            r = requests.get(url, proxies=proxies)  # Добавляем прокси к запросу
            
            # Проверяем код состояния ответа
            if r.status_code == 429:
                retry_after = r.headers.get("Retry-After")
                delay = int(retry_after) if retry_after else 10
                print(f"Слишком много запросов, ждем {delay} секунд.")
                time.sleep(delay)
                continue  # Пропускаем оставшуюся часть цикла и повторяем запрос

            # Если код состояния не 429, то продолжаем обработку страницы
            r.raise_for_status()  # Вызовет исключение для кодов состояния 4xx/5xx
            html = BS(r.content, 'html.parser')
            pages = html.select('.items-items-kAJAg > .iva-item-root-_lk9K')
            
            if not pages:
                break  # Выходим из цикла, если нет контента

            for el in pages:
                info = el.find('div', {'class': 'iva-item-title-py3i_'}).find('a').get('title')
                parts = info.split(', ')
                car_name = parts[0]
                car_age = parts[1]
                car_km = parts[2]
                car_price = parts[-1]

                # Записываем данные в CSV
                writer.writerow([car_name, car_age, car_km, car_price])

            page += 1  # Переходим на следующую страницу

        except requests.exceptions.HTTPError as err:
            print(f"Произошла HTTP ошибка: {err}")
            break  # Прерываем цикл в случае HTTP ошибки
        except requests.exceptions.RequestException as e:
            print(f"Произошла ошибка запроса: {e}")
            break  # Прерываем цикл в случае других ошибок запроса
```

<i> <p style="font-size:20px;"> Определяем верхнюю границу выбросов как Q3 + 1.5*IQR для каждого года

```python
Q1 = df3.groupby('Year')['Price (rub)'].quantile(0.25)
Q3 = df3.groupby('Year')['Price (rub)'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
```

### <i>Фильтруем DataFrame, исключая строки, где цена выше верхней границы выбросов для соответствующего года

```python
df3 = df3[~((df3['Year'].map(upper_bound) < df3['Price (rub)']) & (df3['Year'].map(Q1) > 0))]
```

```python
def create_scatterplot(dataframe, title):
    plt.figure(figsize=(16, 10))
    
    sns.set_style("whitegrid")
    sns.scatterplot(data=dataframe,
                    x='Name',
                    y='Price (rub)',
                    hue='Wear Coefficient',
                    palette='viridis',
                    size='Wear Coefficient',
                    sizes=(20, 200))
    
    plt.xticks(rotation=90, fontsize=10)
    plt.title(title)
    plt.xlabel('Модель автомобиля')
    plt.ylabel('Цена (руб)')
    
    formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6} M')
    plt.gca().yaxis.set_major_formatter(formatter)
    
    plt.legend(title='Wear Coefficient',
               bbox_to_anchor=(1.05, 1),
               loc='upper left')
    
    plt.tight_layout()
    plt.show()
```

```python
create_scatterplot(df_above_median, 'Автомобили с ценой выше медианы')
create_scatterplot(df_below_median, 'Автомобили с ценой ниже медианы')
```

## <u>Отбор топ-10 машин для выгодной покупки
<i> <p style="font-size:18px;">Предполагаем, что чем ниже коэффициент износа и цена за км, и чем выше соотношение цены к возрасту, тем выгоднее покупка

```python
df3.info()
```

```python
df3['Wear Coefficient'] = pd.to_numeric(df3['Wear Coefficient'], errors='coerce')
df3['Price to Age Ratio'] = pd.to_numeric(df3['Price to Age Ratio'], errors='coerce')
df3['Price per km'] = pd.to_numeric(df3['Price per km'], errors='coerce')

df3 = df3.dropna(subset=['Wear Coefficient', 'Price to Age Ratio', 'Price per km'])
```

```python
df3['Value Score'] = 1 / df3['Wear Coefficient'] * df3['Price to Age Ratio'] / df3['Price per km']
top_cars = df3.nlargest(14, 'Value Score')

top_cars['Price per km'].replace(0, 1, inplace=True)
top_cars['Price to Age Ratio'].replace(0, 1, inplace=True)

top_cars = top_cars[(top_cars['Price per km'] > 0) & (top_cars['Price to Age Ratio'] > 0)]

plt.figure(figsize=(34, 16))
scatter = sns.scatterplot(x='Price to Age Ratio',
                          y='Price per km',
                          size='Wear Coefficient',
                          sizes=(100, 1000),
                          hue='Name',
                          data=top_cars,
                          legend=False,
                          palette='viridis')


for line in range(0,top_cars.shape[0]):
    scatter.text(top_cars['Price to Age Ratio'].iloc[line]+0.2, top_cars['Price per km'].iloc[line], 
    top_cars['Name'].iloc[line], horizontalalignment='left', 
    size='medium', color='black', weight='semibold')

    annotation_text = f"{top_cars['Name'].iloc[line]} ({top_cars['Year'].iloc[line]})"
    
    scatter.text(
        x=top_cars['Price to Age Ratio'].iloc[line] + 0.2, 
        y=top_cars['Price per km'].iloc[line], 
        s=annotation_text,
        horizontalalignment='left', 
        size='medium', 
        color='black', 
        weight='semibold'
    )
    
     
        
plt.title('Топ-10 машин для выгодной покупки')
plt.xlabel('Соотношение цены к возрасту')
plt.ylabel('Цена за км')

plt.xscale('log')
plt.yscale('log')


plt.show()
```

<i> <p style="font-size:18px;">Лучшие предложения на рынке относительно возраста/цены: BMW X5 3.0 (2021), BMW 3 серии 2.0 (1993), BMW 5 серии 3.5 (1999), BMW 5 серии 2.5 (1997 и 2005), BMW 5 серии 2.2 (2002).
